```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')
```

```python
mean_area_barrios = inmuebles.groupby('Barrio')['Area'].mean() # agrupar barrio y obtengo promedio del area
mean_area_barrios
```

```python
mean_area_barrios.sort_values(ascending = False, inplace = True) # ordeno valores descendentes
top10_barrios = mean_area_barrios.head(10)
top10_barrios
```

```python
# Grafico top10
top10_barrios.plot.bar(color = 'blue')
plt.title('Top 10 Barrios')
plt.ylabel('Area promedio m2')
plt.xlabel('Barrios')
```

```python
(top10_barrios['Modelia']-top10_barrios['Libertador'])*100/top10_barrios['Modelia']
```

```python
# grafica numero de barrios por UPZ
fig1 = plt.figure()
fig1.set_figwidth(15) # ancho de la figura
barrios_upz.plot.bar(color='blue') # grafico de barras
plt.title('Número de barrios por UPZ')
plt.xlabel('Unidades de Planeacion Zonal')
plt.ylabel('Número de barrios')
```

```python
inmuebles['Tipo'].unique() # que tipo de inmuebles hay
```

```python
# cantidad de inmuebles por tipo por cada UPZ
tipo_upz_count = inmuebles.groupby('UPZ')['Tipo'].value_counts().unstack(level=-1).fillna(0)
tipo_upz_count
```

```python
tipo_upz_count.plot.bar(figsize=(15, 5),stacked =True) # grafico de barras apilado
plt.title("Cantidad de inmuebles a la venta las UPZ")
plt.ylabel("Numero de inmuebles a la venta")
plt.xlabel("Unidades de Planificacion Zonal")
```

```python
# Organizar columna de Valor para ser usada
# reemplazar todo lo que no sean números con '' y eliminar los espacios en blanco
inmuebles['Valor'] = inmuebles['Valor'].replace({'[^0-9]':''}, regex = True).str.strip()
inmuebles['Valor'] = inmuebles['Valor'].astype('int64') # cambiar el tipo de dato a entero
inmuebles.dtypes
```

```python
# promedio valor del inmueble por upz
valor_upz = inmuebles.groupby('UPZ')['Valor'].agg(['min', 'max', 'mean', 'median'])
valor_upz
```

```python
inmuebles.groupby('UPZ')['Valor'].median().plot.bar(figsize=(15, 5))
plt.title("Mediana del Valor del Inmueble")
```
